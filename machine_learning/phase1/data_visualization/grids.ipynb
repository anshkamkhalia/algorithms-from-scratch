```python
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (12,8) # Larger figure size for all plots
sns.set_style("darkgrid")
```

```python
fig, axes = plt.subplots(2,3);


for x in range(0,2):
    for y in range(0,3):
        years = [2020,2021,2022,2023,2024,2025,2026,2027,2028,2029,2030]
        company_a = [random.randint(1,100) for _ in range(len(years))]
        company_b = [random.randint(1,100) for _ in range(len(years))]

        axes[x,y].plot(years, company_a, 'o--r')
        axes[x,y].set_title("Company Earnings")
        axes[x,y].legend(["company_a","company_b"])
        axes[x,y].set_xlabel("Years")
        axes[x,y].set_ylabel("Money Earned (millions)")

# JUST CHANGE AXIS COORDINATES TO MATCH WHICH GRAPH YOU WANT TO DISPLAY IT ON

plt.tight_layout(pad=2)
```
