```python
df = pd.DataFrame({
    "Region": ["Kanto","Johto","Hoenn","Sinnoh","Unova"],
    "Temp": [73,91,87,102,69],
    "Rainfall": [67,88,134,43,96],
    "Humidity": [43,64,58,37,70]
})
```

```python
ktemp = df["Temp"].values[0]
krainfall = df["Rainfall"].values[0]
khumidity = df["Humidity"].values[0]
```

```python
kanto_yield_apples = ktemp*w1+krainfall*w2+khumidity*w3
kanto_yield_apples
```

```python
kanto = [73, 67, 43]
johto = [91, 88, 64]
hoenn = [87, 134, 58]
sinnoh = [102, 43, 37]
unova = [69, 96, 70]
regions = [kanto, johto, hoenn, unova]
```

```python
weights = [w1, w2, w3]
```

```python
def crop_yield(region, weights):
    result = 0
    for x,w in zip(region, weights):
        result += x*w
    return result
```
