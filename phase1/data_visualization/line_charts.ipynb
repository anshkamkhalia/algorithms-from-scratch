```python
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
%matplotlib inline
```

```python
# Set default figure size for all plots (width, height) in inches
plt.rcParams['figure.figsize'] = (12, 8)

# Now every plot will use this size by default
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
example_graph = [random.randint(1,100) for _ in range(10)] # An example graph with random numbers
plt.plot(np.log(example_graph));
```

```python
years = [2020,2021,2022,2023,2024,2025]
example_graph = [random.randint(100000,1000000) for _ in range(len(years))]
plt.plot(years, np.log(example_graph));
plt.xlabel("year"); # Labeling X axis
plt.ylabel("money earned"); # Labeling Y axis
```

```python
# Plotting multiple lines in one chart

years = [2020,2021,2022,2023,2024,2025,2026,2027,2028, 2029, 2030]
company_a = [random.randint(100000,1000000) for _ in range(len(years))]
company_b = [random.randint(100000,1000000) for _ in range(len(years))]

plt.plot(years, company_a);
plt.plot(years, company_b);
plt.xlabel("year");
plt.ylabel("money earned");
```

```python
years = [2020,2021,2022,2023,2024,2025,2026,2027,2028, 2029, 2030]
company_a = [random.randint(100000,1000000) for _ in range(len(years))]
company_b = [random.randint(100000,1000000) for _ in range(len(years))]

# Using marker styles to distinguish points

plt.plot(years, company_a, marker="<");
plt.plot(years, company_b, marker=">");
plt.xlabel("year");
plt.ylabel("money earned");
plt.legend(["company a", "company b"]); # Adding a legend for clarity
```

```python
def generate_data(y, mn, mx):
    """Generates a list of random values for a specificed range"""
    return [random.randint(mn,mx) for _ in range(len(y))]
```

```python
years = [2020,2021,2022,2023,2024,2025,2026,2027,2028, 2029, 2030]

company_a = generate_data(y=years, mn=10000, mx=1000000)
company_b = generate_data(y=years, mn=10000, mx=1000000)

# Uses fmt styling instead of keyword args -> fmt = "[marker][line][color]"

plt.plot(years, company_a, '<--r');
plt.plot(years, company_b, 'o-g');
plt.xlabel("year");
plt.ylabel("money earned");
plt.legend(["company a", "company b"]);
```

```python
years = [2020,2021,2022,2023,2024,2025,2026,2027,2028, 2029, 2030]

company_a = generate_data(y=years, mn=10000, mx=1000000)
company_b = generate_data(y=years, mn=10000, mx=1000000)

plt.plot(years, company_a, '<--r');
plt.plot(years, company_b, 'o-g');
plt.xlabel("year");
plt.ylabel("money earned");
plt.legend(["company a", "company b"]);

# Using a grid
plt.grid(True);
```
