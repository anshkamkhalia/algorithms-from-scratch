```python
plt.title("Sepal Width Distribution");
plt.hist(df.sepal_width); # Creating a histogram
```

```python
plt.hist(df.sepal_width,
         bins=7); # Bins = number of bars
```

```python
plt.hist(df.sepal_width, bins=np.arange(2,5,0.25)); # Using np.arange to make an array that starts at 2 and ends at 5, with a step of 0.25
```

```python
plt.hist(df.sepal_width, bins=np.array([1,3,4,4.5])); # Bins of unequal sizes
```

```python
setosa_df = df[df.species == "setosa"]
versicolor_df = df[df.species == "versicolor"]
virginica_df = df[df.species == "virginica"]
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25)); # Plotting multiple bars on a single chart
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2,5,0.25));
plt.legend(["Setosa", "Versicolor"]);
```

```python
plt.title("Sepal Width Distribution");

plt.hist([setosa_df.sepal_width,
          versicolor_df.sepal_width,
          virginica_df.sepal_width],
         bins=np.arange(2, 5, 0.25),
         alpha=0.5,
         stacked=True); # Stacking histograms

plt.legend(["Setosa", "Versicolor", "Virginica"]);
```
