```python
import matplotlib.pyplot as plt
import seaborn as sns
import random
import numpy as np
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (12,8) # Larger figure size for all plots
sns.set_style("darkgrid")
```

```python
df = sns.load_dataset("flights")
```

```python
df = df.pivot(index="month", columns="year", values="passengers")
```

```python
df
```

```python
plt.title("Number of Passengers (thousands)");
sns.heatmap(df);
```

```python
plt.title("Number of Passengers (thousands)");
sns.heatmap(df, fmt='d', annot=True, cmap='Greens');
```
